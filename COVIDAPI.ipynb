Now we see the data associated with the 'Countries' category. We can pull from that the data we would like based on what is available. 

```python
#Create empty lists. Will store desired data into these lists and convert to data frame. 
country = []
cases = []
deaths = []
date = []
```

#Create DataFrame
search for countries canada and united states. Then extract their country name, total confirmed cases, total deaths, date.

```python
#Import the data for United states and Canada
for i in response['Countries']:
    if i['Slug'] in ('canada', 'united-states'):
        country.append(i['Country'])
        cases.append(i['TotalConfirmed'])
        deaths.append(i['TotalDeaths'])
        date.append(i['Date'])
```

```python
data = pandas.DataFrame({
    'Country': country,
    'cases': cases,
    'deaths': deaths,
    'date': date
})
display(data)
```
